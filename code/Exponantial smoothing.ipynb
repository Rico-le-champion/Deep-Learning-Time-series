# Exponantial smoothing

```python
dim=train.shape
a=np.zeros(dim[0])
```

```python
for i in range (dim[0]):
    data=train.iloc[i,1:].dropna().T.astype('float').to_numpy()
    model = ExponentialSmoothing(data).fit(optimized=True)
    yhat = model.forecast(48)
    a[i]=mean_squared_error(yhat,test.iloc[i,1:].astype('float').T.to_numpy())/np.mean(train.iloc[i,1:].dropna().astype('float').T.to_numpy())
```
